# Remove HTML

```python
from bs4 import BeautifulSoup
import html
```

```python
def remove_html(text):
    text = html.unescape(text)
    return BeautifulSoup(text, "html.parser").get_text(separator=" ")
```

```python
text = remove_html(text)
```

```python
text
```

## Remove URLs and Email Address

```python
import re
```

```python
# Remove URLs
text = re.sub(r'http\S+|www\.\S+', '', text)

# Remove email addresses
text = re.sub(r'\S+@\S+', '', text)
```

# Remove Redundant Letters

```python
def remove_redundant_letters(text):
    return re.sub(r'(.)\1{2,}', r'\1\1', text)

text = remove_redundant_letters(text)
```

# Remove stopwords

```python
# We use NLTK library to remove Stopwords.
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
stopword = stopwords.words('english')
```

```python
# Function
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word not in stopword:
            new_text.append(word)

    return " ".join(new_text)

text = remove_stopwords(text)
```

```python
text = remove_stopwords (text)
```

```python
text
```

# Handling Emoji

## Remove emojis

```python
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

```python
text1 = remove_emoji(text)
```

```python
text1
```

# Remove Punctuation

## Lemmatization

```python
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
import nltk
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag, word_tokenize

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()


def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN


def lemmatize_tokens(tokens):
    tagged = pos_tag(tokens)
    return [lemmatizer.lemmatize(word, get_wordnet_pos(pos)) for word, pos in tagged]
```

```python
tokens = lemmatize_tokens(text)
```

```python
tokens
```

# Remove noise

```python
custom_noise_words = {"omg", "lol", "haha", "wow"}

def remove_custom_noise_words(tokens):
    return [word for word in tokens if word not in custom_noise_words]

tokens = remove_custom_noise_words(tokens)
```
